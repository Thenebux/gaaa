```python
#sorted(iterable, key=funcion, reverse=False)
#Ordena elementos de un iterable.

numeros = [5, 12, 7, 18, 3]
ordenados = sorted(numeros)  
print(ordenados)  # [3, 5, 7, 12, 18]

# Con lambda (ordenar por el último dígito)
ordenados = sorted(numeros, key=lambda x: x % 10)
print(ordenados)  # [12, 3, 5, 7, 18]
```

```python
#Convierte cualquier iterable en lista.
texto = "hola"
print(list(texto))  # ['h', 'o', 'l', 'a']

# Con map
numeros = [1, 2, 3]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)  # [1, 4, 9]
```

```python
#map(funcion, iterable)
#Aplica una función a cada elemento del iterable.

numeros = [1, 2, 3, 4]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)  # [1, 4, 9, 16]
```

```python
#Max, encuentra el valor máximo, opcionalmente con criterio.
palabras = ["pera", "manzana", "uva"]
print(max(palabras))  # uva (alfabéticamente)

# Con lambda (por largo)
print(max(palabras, key=lambda x: len(x)))  # manzana
```

```python
#Igual que max, pero mínimo.
precios = {"manzana": 3, "pera": 2, "uva": 5}
mas_barato = min(precios.items(), key=lambda x: x[1])
print(mas_barato)  # ('pera', 2)
```

```python
inventario = {}

def menu():
    while True:
        print("============[Menu]============")
        print("[1] Agregar producto")
        print("[2] Modificar producto")
        print("[3] Eliminar producto")
        print("[4] Listado en alfabetico")
        print("[5] Valor total del inventario")
        print("[0] Salir")
        print("==============================")
        while True:
            try:
                opcion = int(input("Ingrese una opción: "))
                break
            except(ValueError):
                print("Opcion inválida")
        match opcion:
            case 1: agregar_producto()
            case 2: modificar_producto()
            case 3: eliminar_producto()
            case 4: mostrar_ordenados()
            case 5: valor_total()
            case 0: break
            case _: print("Opcion inválida")

def pedir_dato(mensaje, tipo):
    while True:
        try:
            return tipo(input(mensaje))
        except (ValueError):
            if tipo == int:
                print("Error, el valor debe ser un entero")
            elif tipo == float:
                print("Error, el valor debe ser un real")

def agregar_producto():
    while True:
        codigo = pedir_dato("Ingrese el codigo: ", int)
        if codigo in inventario:
            print(f"Ya existe un producto con ese codigo en el inventario.")
        else: break
    nombre = pedir_dato("Ingrese el nombre: ", str)
    precio = pedir_dato("Ingrese el precio: ", float)
    stock = pedir_dato("Ingrese el stock: ", int)

    inventario[codigo] = (nombre, precio, stock)
    print(f"Producto '{nombre}' ({codigo}) agregado al inventario.")

def modificar_producto():
    while True:
        codigo = pedir_dato("Ingrese el codigo: ", int)
        if codigo not in inventario:
            print(f"El producto no se encuentra en el inventario.")
        else: break
    nombre = pedir_dato("Ingrese el nuevo nombre: ", str)
    precio = pedir_dato("Ingrese el nuevo precio: ", float)
    stock = pedir_dato("Ingrese el nuevo stock: ", int)

    inventario[codigo] = (nombre, precio, stock)
    print(f"Producto '{nombre}' ({codigo}) modificado correctamente.")

def eliminar_producto():
    while True:
        codigo = pedir_dato("Ingrese el codigo: ", int)
        if codigo not in inventario:
            print(f"El producto no se encuentra en el inventario.")
        else: break
    datos = inventario[codigo]
    del inventario[codigo]
    print(f"Producto {datos[0]} ({codigo}) eliminado correctamente.")

def mostrar_ordenados():
    if not inventario:
        print("No hay productos en en el inventario")
        return
    print("Lista de productos ordenados por nombre:")
    for codigo, (nombre, precio, stock) in sorted(inventario.items(), key=lambda item: item[1][0].lower()):
        print(f"Codigo: {codigo} | Nombre: {nombre} | Precio: {precio} | Stock: {stock}")

def valor_total():
    if not inventario:
        print("No hay productos en en el inventario")
        return
    suma=0
    for nombre, precio, stock in inventario.values():
        suma = suma+precio * stock
    print(f"Valor total del inventario: S/.{suma}")

menu()
```
